```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator

import numpy as np
import matplotlib.pyplot as plt

log_path = "/content/drive/MyDrive/LEAP_project/logs/LEAP_model_logs_"
```

## Evaluating the training process:

### Temperature:
- heating tendency

```python
target_name = "heating_tend"

#Path to the TensorBoard log directory
log_dir = log_path+f"{target_name}/version_0"

# Initialize EventAccumulator with the log directory
event_acc = EventAccumulator(log_dir)
event_acc.Reload()

# Get available tags
tags = event_acc.Tags()['scalars']

# Extract training and validation losses
train_loss = []
train_steps = []
val_loss = []
val_step = []

for tag in tags:
    if "train_loss" in tag:
        train_loss = [v.value for v in event_acc.Scalars(tag)]
        train_step = [v.step for v in event_acc.Scalars(tag)]
    elif "val_loss" in tag:
        val_loss = [v.value for v in event_acc.Scalars(tag)]
        val_step = [v.step for v in event_acc.Scalars(tag)]

# Plot the losses over epochs
plt.figure(figsize=(10, 6))
plt.plot(train_step,train_loss, label='Train Loss', color='blue')
plt.plot(val_step,val_loss, label='Validation Loss', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.xticks(np.arange(0,val_step[-1]+98,100000/1024),labels=np.arange(0,len(np.arange(0,val_step[-1]+98,100000/1024)),1))
plt.title(f'Training and Validation Loss over Epochs {target_name}')
plt.legend()
plt.grid(True)
plt.show()
```
